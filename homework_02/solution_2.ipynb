```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("automobile/imports-85.data", header=None, na_values="?")
data.head()
```

```python
data.loc[:, data.dtypes == "float64"] = data.loc[:, data.dtypes == "float64"].fillna(
    data.loc[:, data.dtypes == "float64"].mean(axis=0)
)
data.loc[:, data.dtypes == "object"] = data.loc[:, data.dtypes == "object"].fillna("")
data.head()
```

```python
data_encoded = pd.get_dummies(data)
data_encoded
```

```python
data_encoded.columns = data_encoded.columns.astype(str)
X = data_encoded.drop(columns=["25"])
y = data_encoded["25"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=3)
```

```python
scaler = RobustScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
model = LinearRegression().fit(X_train, y_train)
preds = model.predict(X_test)
```

```python
model = LinearRegression().fit(X_train, y_train)
preds = model.predict(X_test)
print(
    f"""
    MSE_test = {metrics.mean_squared_error(preds, y_test)},
    MAE_test = {metrics.mean_absolute_error(preds, y_test)}
    r2_test = {metrics.r2_score(y_test, preds)}
"""
)
```

Реализуем Хубер лосс, согласно условий из ссылки

```python
class MyLinearRegression:
    def __init__(self, lr=0.005, num_iter=10000, loss_function='mse', delta=1.0) -> None:
        self.lr = lr
        self.num_iter = num_iter
        self.loss_function = loss_function
        self.delta = delta

    def calculate_loss(self, y, y_pred):
        if self.loss_function == 'mse':
            return ((y - y_pred) ** 2).mean()
        elif self.loss_function == 'huber':
            diff = y - y_pred
            abs_diff = np.abs(diff)
            mask = abs_diff <= self.delta
            squared_loss = 0.5 * diff**2
            linear_loss = self.delta * (abs_diff - 0.5 * self.delta)
            return np.where(mask, squared_loss, linear_loss).mean()

    def calculate_dldw(self, y, y_pred, x):
        diff = y - y_pred
        grad_w = 0
        if self.loss_function == 'mse':
            grad_w = (-2 * (x.T * diff).T).mean(axis=0)
        elif self.loss_function == 'huber':
            abs_diff = np.abs(diff)
            mask = abs_diff <= self.delta
            grad_core = np.where(mask, diff, self.delta * np.sign(diff))
            grad_w = (-x.T * grad_core).mean(axis=1)
        return grad_w

    def calculate_dldb(self, y, y_pred):
        diff = y - y_pred
        grad_b = 0
        if self.loss_function == 'mse':
            grad_b = (-2 * diff).mean()
        elif self.loss_function == 'huber':
            abs_diff = np.abs(diff)
            mask = abs_diff <= self.delta
            grad_core = np.where(mask, diff, self.delta * np.sign(diff))
            grad_b = -grad_core.mean()
        return grad_b

    def fit(self, X_train, y_train):
        X_train = np.array(X_train)
        y_train = np.array(y_train)
        self.weights = np.random.rand(X_train.shape[1])
        self.b = 0
        for i in range(self.num_iter):
            pred = (X_train @ self.weights) + self.b
            self.weights -= self.lr * self.calculate_dldw(y_train, pred, X_train)
            self.b -= self.lr * self.calculate_dldb(y_train, pred)

    def predict(self, X_test):
        return X_test @ self.weights + self.b
```

Реализация лин. регрессии с базовыми парамтерами

```python
	
my_model = MyLinearRegression()
my_model.fit(X_train, y_train)
preds = my_model.predict(X_test)
print(
    f"""
    MSE_test = {metrics.mean_squared_error(preds, y_test)},
    MAE_test = {metrics.mean_absolute_error(preds, y_test)}
    r2_test = {metrics.r2_score(y_test, preds)}
"""
)
```

Реализация лин. регрессии с баз. параметрами и Хубер лосс

```python
my_model = MyLinearRegression(lr=0.01, num_iter=10000, loss_function='huber', delta = 1)
my_model.fit(X_train, y_train)
preds = my_model.predict(X_test)
print(
    f"""
    MSE_test = {metrics.mean_squared_error(preds, y_test)},
    MAE_test = {metrics.mean_absolute_error(preds, y_test)}
    r2_test = {metrics.r2_score(y_test, preds)}
"""
)
```

Получили очень большие ошибки MSE и MAE по сравнению с mse реализацией, сверим с встроенной функцией

Результаты стали еще лучше! А что если поставить lr = 0.7 + delta = 10 + num_iter = 1000

```python
my_model = MyLinearRegression(lr=0.7,loss_function='huber', delta = 10, num_iter=1000)
my_model.fit(X_train, y_train)
preds = my_model.predict(X_test)
print(
    f"""
    MSE_test = {metrics.mean_squared_error(preds, y_test)},
    MAE_test = {metrics.mean_absolute_error(preds, y_test)}
    r2_test = {metrics.r2_score(y_test, preds)}
"""
)
```

Стало чуть хуже, но в целом понятно, что для реализации Хубера необходимо подбирать гиперпарамектры. Изменения learning rate позволит начинать обучение с более высокими коэффициентами, но если подобрать num_iter, то даже с исходным получится достичь тех же результтаов, если варьировать delta то так же получится оптимизировать результаты
