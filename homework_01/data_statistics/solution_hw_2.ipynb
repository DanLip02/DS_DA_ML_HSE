 В целом уде получше, далее можно проанализировать выбросы

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound
```

```python
for group, name in [(gate_30_new, 'gate_30'), (gate_40, 'gate_40'), (gate_30_new, 'gate_30_new')]:
    lower, upper = detect_outliers(group, 'sum_gamerounds')
    print(f"\nВыбросы для {name}:")
    print(f"Нижняя граница: {lower:.2f}, Верхняя граница: {upper:.2f}")
    print(f"Доля выбросов: {((group['sum_gamerounds'] < lower) | (group['sum_gamerounds'] > upper)).mean():.2%}")
```

Применим следующие тесты: 
Т-тест для сравнения средних значений.
U-критерий Манна-Уитни для сдвига выборок.
Данные тесты можно применять, так как U-test является непараметричсеким, а T-test для больших выборок ссылается на ЦПТ и средние будут распределены нормально.

```python
from scipy.stats import ttest_ind, mannwhitneyu

t_stat, p_val = ttest_ind(gate_30['sum_gamerounds'], gate_40['sum_gamerounds'], equal_var=False)
print(f"\nРезультаты t-теста:")
print(f"t-статистика: {t_stat:.4f}, p-значение: {p_val:.4f}")
```

```python
u_stat, p_val_u = mannwhitneyu(gate_30['sum_gamerounds'], gate_40['sum_gamerounds'])
print(f"\nРезультаты U-критерия Манна-Уитни:")
print(f"U-статистика: {u_stat:.4f}, p-значение: {p_val_u:.4f}")
```

Для T-test'а принимается H0 о равенстве средних
Для U-test'а так же принимается H0, но очень и очень близкое к пороговому (alpha = 0.05)

```python
def remove_emission(data):
    Q1 = data['sum_gamerounds'].quantile(0.25)
    Q3 = data['sum_gamerounds'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    data_no_outliers = data[(data['sum_gamerounds'] > lower_bound) & (data['sum_gamerounds'] < upper_bound)]
    
    removed_rows = len(data) - len(data_no_outliers)
    print(f"Количество удаленных строк: {removed_rows}")

    return data_no_outliers 
```

Несмотря на то, что KS слишком проваливается и мы отвергаем H0, значение теста очень мало => можно этим принебрегать. В свою же очередь, после удаления выбросов, мы смогли повысить результативность U-test => Тест Манна-Уитни проверяет сдвиг между выборками, и наши выборки в целом не свдинуты друг относительно друга, что показывает и анализ стат. характеристик
В свою же очередь, каждая из подвыборок проходит тест КС в сравнении с изначальным дата-сетом, скорее всего проблема заключается в сдвигах друг относительно друга.

```python
ret_1_table = [[gate_30['retention_1'].sum(), len(gate_30) - gate_30['retention_1'].sum()],
               [gate_40['retention_1'].sum(), len(gate_40) - gate_40['retention_1'].sum()]]

chi2, p, dof, expected = chi2_contingency(ret_1_table)
print(f"Хи-квадрат для retention_1: chi2 = {chi2:.4f}, p-значение = {p:.4f}")
```

```python
ret_7_table = [[gate_30['retention_7'].sum(), len(gate_30) - gate_30['retention_7'].sum()],
               [gate_40['retention_7'].sum(), len(gate_40) - gate_40['retention_7'].sum()]]

chi2, p, dof, expected = chi2_contingency(ret_7_table)
print(f"Хи-квадрат для retention_7: chi2 = {chi2:.4f}, p-значение = {p:.4f}")
```

 Данные расчеты сошлись с Z-test, и довольно хорошо, а занчит можно сделать вывод, что независимо от типа распределения, так как выборка велика, тесты имеют высокую мощность для использования.
